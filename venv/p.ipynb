```python
import pandas as pd
data =(
    {'Name': ['Alice', 'Bob', 'Charlie', 'David','Eve','liza','sara','zen','tara'], 
        'Age': [24, 27, 22, 32,29,21,23,25,26],
        'Salary': [50000, 60000, 55000, 70000,65000,48000,52000,58000,62000],
    }
)

df=pd.DataFrame(data)   

print(df.head(3))  # Display the first 3 rows of the DataFrame
print(df.tail(2))  # Display the last 2 rows of the DataFrame
print(df.shape)  # Display summary statistics of the DataFrame
print(df.info())   # Display information about the DataFrame

df['Bonus']=df['Salary']*0.10
df['Salary'] = df['Salary'] + 5000
print(df) 
```
